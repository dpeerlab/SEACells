```python
# grab the file results10/10000_cells/sparsity_v4_1701760643.2466624.csv
# and plot the results

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# read in the data
df = pd.read_csv('sparsity_v4_1701760643.2466624.csv')

# Drop columns Unnamed: 0, B, and K 
df = df.drop(['Unnamed: 0', 'B', 'K'], axis=1)
```

```python
# Make a scatter plot of the data 
sns.scatterplot(data = df, legend=False)
sns.lineplot(data = df, legend=False)
# SEt title 
plt.title('Sparsity of A vs. Iteration')
plt.xlabel('Iteration')
plt.ylabel('Sparsity of A')
```
